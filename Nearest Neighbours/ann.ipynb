```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
#usig make_blobs to create a dataset of 300 samples, 3 centers, and 2 features
X,y = make_blobs(n_samples=300,
                 centers=3,
                 random_state=0,
                 cluster_std=0.60,
                 n_features=2)

plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='viridis')
plt.show()
```

```python
# create a new point to pedict its class
new_point = np.random.uniform(low=-3,high=3,size=(1,2))

plt.scatter(new_point[:,0],new_point[:,1],c='red',s=100,marker='X')
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='viridis')
```

```python
dist = X[:, :2] - new_point        # (300,2)
distance = np.linalg.norm(dist, axis=1)   # (300,)

k = 5
k_nearest = np.argsort(distance)[:k]
k_nearest_distances = distance[k_nearest]

print(k_nearest)
print(k_nearest_distances)

pred = np.bincount(y[k_nearest]).argmax()
print(pred)
```
